```python
# Group crops into broader categories
crop_groups = {
    'Pulses': ['Arhar/Tur', 'Gram', 'Moong(Green Gram)', 'Urad', 'Other Rabi pulses', 'Other Kharif pulses', 'Peas & beans (Pulses)', 'Other Summer Pulses', 'Moth','Cowpea(Lobia)',],
    'Oilseeds': ['Castor seed', 'Linseed', 'Rapeseed &Mustard', 'Sesamum', 'Niger seed', 'Groundnut', 'Soyabean', 'Safflower', 'Sunflower', 'other oilseeds', 'Oilseeds total'],
    'Cereals': ['Rice', 'Wheat', 'Maize', 'Barley', 'Jowar', 'Bajra', 'Ragi', 'Small millets', 'Other Cereals'],
    'Spices': ['Dry chillies', 'Black pepper', 'Cardamom', 'Coriander','Turmeric'],
    'Fruits': ['Banana', 'Cashewnut'],
    'Tubers': ['Potato', 'Sweet potato', 'Tapioca'],
    'Fibers': ['Cotton(lint)', 'Jute', 'Mesta'],
    'Vegetables': ['Onion', 'Khesari', 'Sannhamp', 'Guar seed','Garlic', 'Ginger'],
    'Sugarcane':['Sugarcane']
}

# Function to map crops to their group
def map_crop_to_group(crop):
    for group, crops in crop_groups.items():
        if crop in crops:
            return group
    return 'Miscellaneous'

# Apply the mapping function
ag['Crop'] = ag['Crop'].apply(map_crop_to_group)

print(ag['Crop'].value_counts())



# Ye code use kar lena lekin Pehle 'Coconut' ko hata dena data se
```

```python
# Define a function to categorize the crop years into decades
def categorize_decade(year):
    if 1997 <= year <= 2006:
        return '1997-2006'
    elif 2007 <= year <= 2016:
        return '2007-2016'
    elif 2017 <= year <= 2020:
        return '2017-2020'

# Apply the function to create a new 'Decade' column
assam['Decade'] = assam['Crop_Year'].apply(categorize_decade)

# Group by 'Decade' and calculate the total or average yield
decade_grouped = assam.groupby('Decade')['Yield(tons/ha)'].mean().reset_index()

# Plotting the bar plot
plt.figure(figsize=(8, 5))
plt.bar(decade_grouped['Decade'], decade_grouped['Yield(tons/ha)'], color=['skyblue', 'lightgreen', 'salmon'])
plt.title('Average Yield per Decade')
plt.xlabel('Decade')
plt.ylabel('Average Yield (tons/ha)')
plt.show()
```

```python
# Define a function to categorize the crop years into decades
def categorize_decade(year):
    if 1997 <= year <= 2006:
        return '1997-2006'
    elif 2007 <= year <= 2016:
        return '2007-2016'
    elif 2017 <= year <= 2020:
        return '2017-2020'

# Apply the function to create a new 'Decade' column
assam['Decadee'] = assam['Crop_Year'].apply(categorize_decade)

# Group by 'Decade' and calculate the necessary aggregates
decade_grouped = assam.groupby('Decadee').agg({
    'Yield(tons/ha)': 'mean',               # Average yield
    'Annual_Rainfall(mm)': 'sum',                # Total annual rainfall
    'Fertilizer(kg)': 'sum',                     # Total fertilizer usage
    'Pesticide(kg)': 'sum'                       # Total pesticide usage
}).reset_index()

# Combine fertilizer and pesticide sums
decade_grouped['Fertilizer(kg) + Pesticide(kg)'] = decade_grouped['Fertilizer(kg)'] + decade_grouped['Pesticide(kg)']

# Plotting the bar plot with multiple bars
bar_width = 0.2
index = np.arange(len(decade_grouped['Decadee']))

plt.figure(figsize=(10, 6))

# Yield bar
plt.bar(index, decade_grouped['Yield(tons/ha)'], bar_width, label='Average Yield')

# Annual Rainfall bar
plt.bar(index + bar_width, decade_grouped['Annual_Rainfall(mm)'], bar_width, label='Total Annual Rainfall')

# Fertilizer + Pesticide bar
plt.bar(index + 2 * bar_width, decade_grouped['Fertilizer(kg) + Pesticide(kg)'], bar_width, label='Fertilizer(kg) + Pesticide(kg)')

# Customizing the plot
plt.xlabel('Decadee')
plt.ylabel('Values')
plt.title('Decade-wise Comparison of Yield, Rainfall, and Fertilizer(kg) + Pesticide(kg)')
plt.xticks(index + bar_width, decade_grouped['Decadee'])
plt.legend()

plt.show()
```
